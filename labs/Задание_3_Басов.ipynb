```python
import pandas as pd
from numpy import nan

df = pd.read_csv('/content/ca_offenses_by_county.csv')
df.info()

# Преобразуем типы
df['Metropolitan/Nonmetropolitan'] = df['Metropolitan/Nonmetropolitan'].astype("category")
df['County'] = df['County'].astype("category")

columns = ['Violent crime', 'Murder and nonnegligent manslaughter', 'Rape (revised definition)', 'Robbery', 'Aggravated assault', 'Property crime', 'Burglary', 'Larceny-theft', 'Motor vehicle theft', 'Arson']

for column_name in columns:
  df[column_name] = df[column_name].astype(str).str.replace('[^0-9.]', '', regex=True).replace('', nan).astype(float)

print(f'\n{df.dtypes}')
```

```python
df.isnull().sum()
```

Столбец "Rape (legacy definition)" полностью состоит из NaN. Убираем.

```python
df = df.dropna(subset=[5], axis=1)
```

```python
#Удаление

df_delete = df.copy()
df_delete = df_delete.dropna(how='all')
df_delete
```

```python
df_delete.isnull().sum()
```

```python
#Заполнение при помощи медианы
df_filled_median = df.copy()
df_columns = ['Violent crime', 'Murder and nonnegligent manslaughter', 'Rape (revised definition)', 'Robbery', 'Aggravated assault', 'Property crime', 'Burglary', 'Larceny-theft', 'Motor vehicle theft', 'Arson']

for column_name in df_columns:
  df_filled_median[column_name].fillna(df_filled_median[column_name].median(), inplace=True)

df_filled_median
```

```python
#Заполнение при помощи среднего
df_filled_mean = df.copy()
df_columns = ['Violent crime', 'Murder and nonnegligent manslaughter', 'Rape (revised definition)', 'Robbery', 'Aggravated assault', 'Property crime', 'Burglary', 'Larceny-theft', 'Motor vehicle theft', 'Arson']

for column_name in df_columns:
  df_filled_mean[column_name].fillna(df_filled_mean[column_name].mean().round(1), inplace=True)

df_filled_mean
```

```python
#Заполнение при помощи моды
df_filled_mode = df.copy()
df_columns = ['Violent crime', 'Murder and nonnegligent manslaughter', 'Rape (revised definition)', 'Robbery', 'Aggravated assault', 'Property crime', 'Burglary', 'Larceny-theft', 'Motor vehicle theft', 'Arson']

for column_name in df_columns:
  df_filled_mode[column_name].fillna(df_filled_mode[column_name].mode()[0], inplace=True)

df_filled_mode
```

```python
#Заполнение при помощи интерполяции
df_filled_interpol = df.copy()
df_columns = ['Violent crime', 'Murder and nonnegligent manslaughter', 'Rape (revised definition)', 'Robbery', 'Aggravated assault', 'Property crime', 'Burglary', 'Larceny-theft', 'Motor vehicle theft', 'Arson']

for column_name in df_columns:
  df_filled_interpol[column_name].fillna(df_filled_interpol[column_name].interpolate(), inplace=True)

df_filled_interpol
```

###Пример реализации

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```

```python
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}
df = pd.DataFrame(data)
df
```

```python
corr_pearson(df["Рост"], df["Вес"])
```
